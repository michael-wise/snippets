```python
import librosa, librosa.display
import matplotlib.pyplot as plt  
import numpy as np
```

```python
y, sr = librosa.load("C:\\Users\\weezi\\Downloads\\Interrupted_By_Fireworks_-_An_Orchestral_Tribute_t.wav")
S = np.abs(librosa.stft(y)**2)
chroma = librosa.feature.chroma_stft(S=S, sr=sr)
```

```python
plt.figure()
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chromagram: Interrupted by fireworks')
```

```python
plt.show()
```

```python
#https://librosa.github.io/librosa/generated/librosa.segment.recurrence_matrix.html#librosa.segment.recurrence_matrix
#Find nearest neighbors in MFCC space
plt.figure(figsize=(8, 4))
mfcc = librosa.feature.mfcc(y=y, sr=sr)

R = librosa.segment.recurrence_matrix(mfcc)
R_aff = librosa.segment.recurrence_matrix(mfcc, mode='affinity')

librosa.display.specshow(R, x_axis='time', y_axis='time')

plt.title('Binary recurrence (symmetric)')
plt.subplot(1, 2, 2)
librosa.display.specshow(R_aff, x_axis='time', y_axis='time', cmap='magma_r')
plt.title('Affinity recurrence')
plt.tight_layout()
```
