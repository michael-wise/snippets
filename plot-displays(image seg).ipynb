```python
#Gets bounding box (bbox) of SNES window.
import win32gui, pywinauto, re, numpy
import pyscreenshot as ImageGrab
pat = re.compile('.*Snes9X.*')
hwnd = pywinauto.findwindows.find_windows(title_re=pat)
snesbbox = win32gui.GetWindowRect(hwnd[0])
print(snesbbox)
```

```python
im = ImageGrab.grab(bbox=snesbbox)
plt.imshow(im)
plt.
plt.show()
```

```python
fig = plt.figure();
ax = fig.gca();
ax.set_xticks(numpy.arange(0, snesbbox[-2], snesbbox[-2]/10))
ax.set_yticks(numpy.arange(0, snesbbox[-1], snesbbox[-1]/10))
ax.patch.set_alpha(0)
plt.grid()
plt.imshow(im)
plt.show()
```
